# 1. Create dataset

```python
from sklearn.datasets import make_moons
X, y = make_moons(n_samples=10000, noise=0.4, random_state=112)
```

# 2. Split train, validation

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.2, random_state=42)
```

# 3. Find the best tree model

```python
from sklearn.tree import DecisionTreeClassifier
tree_clf = DecisionTreeClassifier()
```

```python
from sklearn.model_selection import GridSearchCV

# max_depth=None,
# min_samples_split=2,
# min_samples_leaf=1,
# min_weight_fraction_leaf=0.0,
# max_features=None
grid_params = [
    {
        'max_depth': [1, 2, 3, 10, 100, 300],
        'min_samples_split': [10, 50, 100, 300, 500],
        'min_samples_leaf': [10, 50, 100, 300, 500],
        'min_weight_fraction_leaf': [.001, .01, .05, 0.1, 0.2, 0.4],
    }
]
grid_search = GridSearchCV(tree_clf, grid_params, cv=5)
grid_search.fit(X_train, y_train)
```

```python
import numpy as np
cvres = grid_search.cv_results_
for mean_score, std_score, params in zip(cvres['mean_test_score'], cvres['std_test_score'], cvres['params']):
    print(np.sqrt(mean_score), np.sqrt(std_score), params)
```

# 4. Check on validatinon set

```python
best_tree_clf = grid_search.best_estimator_
y_predict = best_tree_clf.predict(X_test)

from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_predict) * 100
```
